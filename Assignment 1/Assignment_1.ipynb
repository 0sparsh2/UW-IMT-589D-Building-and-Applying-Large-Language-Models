```python
import torch

# Open the file in read mode
with open('/content/anne.txt', 'r') as file:
  content = file.read().lower()
  words = content.split()
```

```python
words = [word for word in words if word.isalpha() and 'Ã¦' not in word]
```

```python
# Now let's look at the trigram model. We need a way to count.
tm = {}
for w in words:
  str = ['.','.'] + list(w) + ['.']
  for ch1, ch2, ch3 in zip(str, str[1:], str[2:]):
    trigram = (ch1, ch2, ch3)
    tm[trigram] = tm.get(trigram, 0) + 1 # If the trigram doesn't exist, assign 0
tm
```

```python
tm.items()
```

```python
sorted(tm.items(), key = lambda kv: -kv[1])
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
stoi
```

```python
itos = {i:s for s, i in stoi.items()}
itos
```

```python
N = torch.zeros((27,27,27), dtype=torch.int32)
for w in words:
  str = ['.','.'] + list(w) + ['.']
  for ch1, ch2, ch3 in zip(str, str[1:], str[2:]):
    index1 = stoi[ch1]
    index2 = stoi[ch2]
    index3 = stoi[ch3]
    N[index1,index2,index3] += 1
N
```

```python
# We are trying to generate words
g = torch.Generator().manual_seed(1234567890)

for i in range(20):
  out = []
  index1 = 0
  index2 = 0
  while True:
    p = N[index1, index2].float()
    p=p/p.sum()
    index3 = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
    out.append(itos[index3])
    index1 = index2
    index2 = index3
    if index3 == 0:
      break



  print(''.join(out))
```
